# How to sort pandas dataframe from by one column ?

### Scenario

```python
import pandas as pd
month_name = ['February','August',  'April',  'December',    'January',
 'July', 'June',  'March',  'May', 'November', 'October', 'September']
month_number = [2,8,4,12,1,7,6,3,5,11,10,9]
months_df = pd.DataFrame({"month_name":month_name,
                          "month_number":month_number})
months_df
```

## Sorting on Numerical Column

```python
months_df.sort_values("month_number") #ascending=True
```

```python
months_df.sort_values("month_number",ascending=True) #
```

```python
months_df.sort_values("month_number",ascending=False,inplace=True)
months_df
```

## Sorting on a categorical column

```python
months_df.sort_values("month_name")
```

```python
months_df.sort_values("month_name",ascending=False)
```
