```python
mtcars.gear.value_counts().reset_index()
```

```python
mtcars.groupby("gear").size().reset_index(name="num_gears")
```

```python
mtcars.groupby(["cyl","am"]).size().reset_index(name="num_gears")
```

```python
am_count = mtcars.am.value_counts().reset_index()
am_count.columns = ["am", 'count'] 
am_count
```

```python
am_gear_group = mtcars.groupby(["am", "gear"])
am_gear_group.size()
```

```python
mtcars.groupby(["am", "gear"]).size().reset_index().rename(columns={0:"count"})
```

```python
mtcars.groupby(["am", "gear"])\
.size()\
.reset_index()\
.rename(columns={0:"count"})
```

```python
mtcars.groupby(["am"]).agg({"am":"count"})
```
