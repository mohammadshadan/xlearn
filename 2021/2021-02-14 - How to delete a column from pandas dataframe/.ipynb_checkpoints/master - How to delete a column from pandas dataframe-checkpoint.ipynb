# Python - How to delete a column from pandas dataframe ?

```python
import numpy as np
import pandas as pd
```

## Creating Scenario

```python
df = pd.DataFrame({'mpg': {0: 21.0, 1: 21.0, 2: 22.8, 3: 21.4, 4: 18.7},
 'cyl': {0: 6, 1: 6, 2: 4, 3: 6, 4: 8},
 'disp': {0: 160, 1: 160, 2: 108, 3: 258, 4: 360},
 'hp': {0: 110, 1: 110, 2: 93, 3: 110, 4: 175},
 'wt': {0: 2.62, 1: 2.875, 2: 2.32, 3: 3.215, 4: 3.44},
 'vs': {0: 0, 1: 0, 2: 1, 3: 1, 4: 0},
 'am': {0: 1, 1: 1, 2: 1, 3: 0, 4: 0},
 'gear': {0: 4, 1: 4, 2: 4, 3: 3, 4: 3}})
df
```

```python
# Expected Output
```

## Method 1

```python
df = df.drop("gear", axis = 1)
df
```

```python
df.drop("gear", axis = 1, inplace = True)
df
```

# Method 2

```python
df1 = df.drop(columns="gear")
df1
```

```python
df.drop(columns="gear", inplace = True)
```

## Summary 
To drop a single from pandas dataframe
- df.drop("column_name", axis=1)
- df.drop("column_name", axis = 1, inplace = True ) #rewrites the df datarame itself
- df.drop (columns = "column_name") #no need to mention the axis
- df.drop(columns = "column_name", inplace = True)
